Ej.2 - Calcula el perímetro y área de un rectángulo, dada su base y su altura.

```python
base = float(input("Introduce la base del rectángulo: "))
altura = float(input("Introduce la altura del rectángulo: "))

area = base * altura
perimetro = 2 * (base + altura)

print(f"El área del rectángulo es:", area)
print(f"El perímetro del rectángulo es:", perimetro)
```

Ej.3 - Calcula el perímetro y área de una circunferencia, dado su radio. Importar con import
math y utilizar la librería math para acceder al valor de la constante math.pi.

```python
import math

radio = float(input ("Introduce el radio de la circunferencia: "))

area = math.pi * radio ** 2
perimetro = 2 * math.pi * radio

print(f"El área de la circunferencia es:", area)
print(f"El perímetro de la circunferencia es:", perimetro)
```

Ej. 15 - Crea una aplicación que pida un número y calcule su factorial. El factorial de un
número es el producto de todos los números enteros entre 1 y el ese número y se representa por el número seguido de un signo de exclamación. P. ej. 6! = 1x2x3x4x5x6=720. Realiza una
propuesta con while y otra con for.

```python
#Factorial

numero = input("Introduce un numero para calcular su factorial: ")
numero = int(numero)
factorial = 1

for i in range(1, numero+1):
    factorial *= i

print("El factorial de", numero, "es", factorial)




```

Ej. 16 - Crea una aplicación que permita adivinar un número. En primer lugar, la aplicación
solicita un número entero por teclado, que será el número que debe adivinarse. A
continuación, va pidiendo números y va indicando por pantalla, si el número a adivinar es
mayor o menor que el introducido. El programa termina cuando se acierta el número.

```python
numero_secreto = input("Introduce un numero para adivinar: ")
numero_secreto = int(numero_secreto)

while True:
    numero_usuario = input("Intenta adivinar el numero: ")
    numero_usuario = int(numero_usuario)

    if numero_usuario == numero_secreto:
        print("¡Adivinaste!")
        break
    else:
      print("Adivina de nuevo")



```

Ej.18 - Escribe un programa que diga si un número introducido por teclado es o no primo. Un
número primo es aquel que sólo es divisible entre él mismo y la unidad.

```python
numero = input("Introduce un numero para saber si es primo: ")
numero = int(numero)
es_primo = True

for i in range(2, numero):
    if numero % i == 0:
        es_primo = False
        break

if es_primo:
    print("El numero es primo")
else:
    print("El numero no es primo")
```

Ej.7 - Dada una lista de cadenas, pide una cadena por teclado e indica si dicha cadena está
en la lista. También indica cuantas veces aparece esa cadena en la lista. Lee otra cadena y
la sustituye por la primera.

```python
lista=["rojo", "azul", "verde", "blanco", "rojo", "amarillo", "rojo", "verde"]

cadena = input("Intrduce una cadena de texto")

contador = lista.count(cadena) #contador para saber cuantas veces esta en la lista

if contador > 0:
    print ("La palabra " ,cadena, " esta en la lista y esta " ,contador, " veces.")

    cadena2 = input("Introduce otra cadena de texto")
    lista[0] = cadena2 #sustituir la palabra por la primera en la lista
    print (lista)
else:
    print("La palabra" ,cadena,"no esta en la lista.")
```

Ej.8 - Dada una lista de números, el programa debe indicar si la lista está ordenada o no.

```python
numeros = [1,5,7,2,6,9,34,134,2435,1,3]

numerosOrdenados=numeros.copy()
numerosOrdenados.sort()

if numerosOrdenados == numeros:
    print ("La lista esta ordenada")
else:
    print("La lista no esta ordenada")
```

Ej.9 - Escribe un programa que lea una cadena por teclado, y devuelva un diccionario con la
cantidad de apariciones de cada palabra.

```python
cadena = input("Introduce una cadena de texto")
palabras= cadena.split()

contador = {}

for palabra in palabras:
    if palabra in contador:
        contador[palabra] += 1
    else:
        contador[palabra] = 1
print(contador)
```

Ej.13 - Escribe un programa que gestione una estructura donde tengamos almacenados
personas y lista de gustos de personas. Deberá pedir por teclado una persona y algo que le
guste. Si la pareja persona y gusto no existe debe añadirlos. Si existe la persona, pero no el
gusto, debe añadir el gusto, y si ya existen no hace nada. Parará de pedir persona y gusto
cuando se introduzca 0.

```python
personas = {}

while True:
    nombre = input("Introduce tu nombre, (0 para salir)").strip()

    if nombre == "0":
        break

    gusto = input("Introduce tu hobby").strip()


    if nombre not in personas:
        personas[nombre] = [gusto]
    else:

        if gusto not in personas[nombre]:
            personas[nombre].append(gusto)
        else:
            print(nombre,"ya tiene", gusto, "en su lista")

    print(personas)
```
